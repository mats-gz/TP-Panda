```python
import pandas as pd

#Traer el dataset como variable df
df = pd.read_csv("./Mobiles Dataset (2025).csv", encoding="ISO-8859-1")

#Realizar el promedio de celulares lanzados en 2024
year1_average = len(df[df["Launched Year"] == 2023])
print("La cantidad de celulares fabricados en 2023 son:", year1_average)

#Filtrar los celulares de Apple y dar el promedio de su peso
filter_apple = df[df["Company Name"] == "Apple"]
weight_average = filter_apple["Mobile Weight"].str.replace("g"," ").astype(float).mean()
print("El promedio del peso de los celulares Apple es:", weight_average)

#Filtrar la RAM de los celulares mayores a 800 USD para ver cual es la RAM m√°s com√∫n
filtered_price = df[df["Launched Price (USA)"] > "USD 800"]
ram = filtered_price["RAM"].str.replace("GB"," ").astype(float).mode()
print("La RAM m√°s com√∫n de los celulares con precio mayor a 800 USD es:", ram.iloc[0], "GB, una explicaci√≥n razonable de su alto costo.\n")

#Agrupar los nombres de las empresas, para ver el promedio de sus precios de celular y ver cual de todas vende sus celulares m√°s caros que la competencia
df["Launched Price (USA)"] = (
    df["Launched Price (USA)"]
    .str.replace("USD", "")
    .str.replace(",", "")
    .str.strip()
    .astype(float)
)
group_apple = df.groupby(df["Company Name"])["Launched Price (USA)"].mean()
print(group_apple)
print("En base a la lista proporcionada, podemos ver que Apple, Sony y Huawei poseen los precios m√°s elevados en la industria de tel√©fonos m√≥viles.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Traer el dataset como variable df
df = pd.read_csv("./Mobiles Dataset (2025).csv", encoding="ISO-8859-1")

#Agrupar Apple y sus a√±os de trayectoria, para ver el aumento de sus precios en los a√±os de empresa
df["Launched Price (USA)"] = (
    df["Launched Price (USA)"]
    .str.replace("USD", "")
    .str.replace(",", "")
    .str.strip()
    .astype(float)
)
apple = df[df["Company Name"] == "Apple"]
apple_years = apple.groupby("Launched Year")["Launched Price (USA)"].mean()

#Agrupar los tama√±os de pantalla de los celulares con sus precios de 2023, y observar las tendencias del mercado. ¬øEs una pantalla m√°s grande, un precio m√°s elevado?
df["Screen Size"] = (
    df["Screen Size"]    
    .str.replace("inches", "in.")
)
year_2023 = df[df["Launched Year"] == 2023]
screen_2023 =  year_2023.groupby("Screen Size")["Launched Price (USA)"].mean()
screen_2023 = screen_2023.tail(14)

plt.figure(figsize=(8, 4))
plt.bar(apple_years.index, apple_years.values, label="A√±os de Precios - Apple", color="pink")
plt.xlabel("A√±os")
plt.ylabel("Promedio de precios")
plt.title("Promedio de precios de Apple 2017-2024")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 4))
plt.scatter(screen_2023.index, screen_2023.values, label="Tama√±os de Pantalla - Precios", color="grey")
plt.xlabel("Tama√±o de Pantala")
plt.ylabel("Promedio de precios")
plt.title("Relaci√≥n precios - Pantalla")
plt.legend()
plt.grid()
plt.show()
```

### ÀñÀö‚äπ Í£ë‡ßé - üíå. Analisis de Gr√°ficos 
#####  ö…û - En el primer gr√°fico podemos observar la diferencia de promedio de precios de Apple. Si bien puede parecer un gr√°fico simple, este muestra a√±os cr√≠ticos de dicha empresa, debido a que, el promedio, se ve afectado por la cantidad de productos que sacan por a√±o. Apple, generalmente, lanza 12 productos por a√±o (con respecto a celulares), pero en 2019 sac√≥ solo 9 de los mismos, debido a que la empresa pasaba por un momento cr√≠tico por la disputa comercial entre USA y China, lo que afect√≥ su cadena de suministro y precios. La compa√±√≠a experiment√≥ una ca√≠da significativa en su valor de mercado y acciones, principalmente debido a la amenaza de aranceles sobre sus productos. Esto es debido a que Apple depend√≠a en gran medida de China para la fabricaci√≥n de sus productos, afectando tanto sus precios como su cantidad de productos lanzados en dicho a√±o.
##### êôö - En el segundo gr√°fico observamos la relaci√≥n entre el tama√±o de pantalla de los celulares lanzados en 2023 y su precio promedio. Aunque puede parecer que una pantalla m√°s grande implica un precio m√°s elevado, los datos muestran que no siempre es as√≠. Algunos celulares con pantallas grandes tienen precios bajos, mientras que otros con tama√±os m√°s normales (alrededor de 6.8 in. a 7.6 in.) alcanzan precios mucho m√°s altos. Esto se puede deber a que, adem√°s del tama√±o, influye mucho la marca, las caracter√≠sticas del celular y si es un modelo plegable o "premium".
